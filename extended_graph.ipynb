```python
import xarray as xr
import pandas as pd
```

```python
dataset = xr.open_dataset("/Volumes/EB406_T7_2/new_routing/Qout_new.nc")
```

```python
subset = dataset.sel(rivid=160453644)
```

```python
subset
```

```python
import matplotlib.pyplot as plt

# Ensure 'time' is a coordinate and 'Q' is the data variable
subset = dataset.sel(rivid=160453644)

# Plot each ensemble member on the same plot
plt.figure(figsize=(10, 6))
for ensemble_member in subset['ensemble']:
    plt.plot(subset['time'], subset['Q'].sel(ensemble=ensemble_member), label=f'Ensemble {ensemble_member.values}')

# Customize the plot
plt.xlabel('Time')
plt.ylabel('Q')
plt.title('Q over Time for Each Ensemble Member')
plt.tight_layout()
plt.savefig('/Users/rachel1/Downloads/high_resolution_for_thesis/ensemble_plot_160453644.png', dpi=600, bbox_inches='tight')
plt.show()
```
